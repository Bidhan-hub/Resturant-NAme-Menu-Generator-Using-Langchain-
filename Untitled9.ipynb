```python
import os
from langchain_openai import OpenAI

# Set OpenAI API key
os.environ["OPENAI_API_KEY"] = my_sk

llm = OpenAI(
    temperature=0.6,
    model="gpt-3.5-turbo-instruct"  # IMPORTANT: not gpt-3.5-turbo
)



```

```python
from langchain_core.prompts import PromptTemplate       # with the help of prompt Templete we are able to get the Resturant Name by Inserting the Cuisine.
prompt_template_name = PromptTemplate(
    input_variables=["cuisine"],
    template="Suggest a fancy restaurant name for {cuisine} food."
)

prompt = prompt_template_name.format(cuisine="American")

response = llm.invoke(prompt)
print(response)
```

```python
## LLMChain for Restaurant Name
name_chain = prompt_template_name | llm

result = name_chain.invoke({"cuisine": "Nepali"})
print(result)
```

```python
# Menu prompt
prompt_template_menu = PromptTemplate(
    input_variables=["cuisine"],
    template="Create a 5-item premium menu for a {cuisine} restaurant."
)

# Runnable chains
name_chain = prompt_template_name | llm
menu_chain = prompt_template_menu | llm

### Invoke chains
name = name_chain.invoke({"cuisine": "Nepali"})
menu = menu_chain.invoke({"cuisine": "Nepali"})

```

```python
print("Restaurant Name:\n", name)
print("\nMenu:\n", menu)
```

```python
###Streamlit POC Application
import os
import streamlit as st
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI

os.environ["OPENAI_API_KEY"] = my_sk

st.set_page_config(page_title="Restaurant Idea Generator", page_icon="üçΩÔ∏è")
st.title("üçΩÔ∏è Restaurant Idea Generator")

cuisine = st.text_input("Enter Cuisine (e.g., Nepali, Italian, Indian)")

llm = OpenAI(
    temperature=0.6,
    model="gpt-3.5-turbo-instruct"
)

prompt = PromptTemplate(
    input_variables=["cuisine"],
    template="""
    Create:
    1. A fancy restaurant name
    2. A tagline
    3. 5 premium menu items
    for {cuisine} cuisine.
    """
)

# Runnable chain
chain = prompt | llm

if cuisine:
    response = chain.invoke({"cuisine": cuisine})
    st.subheader("‚ú® Generated Idea")
    st.write(response)
```
